Dataset: IBM HR Analytics (1,470 employees, 35 features)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid', palette='muted')

df = pd.read_csv('../data/HR-Employee-Attrition.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,4))
df['Attrition'].value_counts().plot(kind='bar', ax=axes[0], color=['#01696f','#e5534b'], rot=0)
axes[0].set_title('Attrition Distribution')
df.groupby('Attrition')['Age'].plot(kind='kde', ax=axes[1], legend=True)
axes[1].set_title('Age Distribution by Attrition')
plt.tight_layout(); plt.show()
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.select_dtypes(include='number').corr(), cmap='coolwarm', center=0, linewidths=.3)
plt.title('Feature Correlation Matrix')
plt.tight_layout(); plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15,8))
for ax, feat in zip(axes.flat, ['OverTime','JobSatisfaction','WorkLifeBalance','MonthlyIncome','YearsAtCompany','DistanceFromHome']):
    if df[feat].dtype=='object':
        sns.countplot(x=feat, hue='Attrition', data=df, ax=ax)
    else:
        sns.boxplot(x='Attrition', y=feat, data=df, ax=ax)
    ax.set_title(feat)
plt.suptitle('Key Attrition Drivers', fontsize=14)
plt.tight_layout(); plt.show()
```

## Key Findings
- **Overtime** employees show 3x higher attrition rates
- **Monthly income < $3,000** is the strongest single predictor
- **Low job satisfaction (1-2)** doubles attrition likelihood
- **< 3 years at company** is peak attrition risk period
